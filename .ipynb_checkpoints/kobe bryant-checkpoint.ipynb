```python
data = pd.read_csv("data.csv")
```

```python
data.columns
```

```python
# Visualization of shots, misses and made
court_scale, alpha = 7, 0.05
plt.figure(figsize=(2 * court_scale, court_scale*(84.0/50.0)))
plt.subplot(121)
h = data.loc[data.shot_made_flag == 1]
plt.scatter(h.loc_x, h.loc_y, color='blue', alpha=alpha)
plt.title('Shots Made')
ax = plt.gca()
ax.set_ylim([-50, 900])
plt.subplot(122)
h = data.loc[data.shot_made_flag == 0]
plt.scatter(h.loc_x, h.loc_y, color='red', alpha=alpha)
plt.title('Shots missed')
ax = plt.gca()
ax.set_ylim([-50, 900])
plt.savefig('shots_made_and_missed.png')
```

```python
groups = data.groupby('combined_shot_type')
fig, ax = plt.subplots(figsize=(court_scale, court_scale*(84.0/50.0)))
alpha = 0.2
alphas, n = [], float(len(data.combined_shot_type))
for u in [i[0] for i in groups]:
    d = len(data.loc[data.combined_shot_type == u, 'combined_shot_type'])
    alphas.append(np.log1p(d))
colors = ['blue', 'red', 'orange', 'green', 'yellow', 'pink']
for (name, group), alp in zip(groups, alphas):
    ax.plot(group.loc_x, group.loc_y,
            marker='.', linestyle='', ms=12,
            label=name, alpha=alp)
plt.title("Shot Type")
ax.legend()
plt.savefig('combined_shot_type_layout.png')
```

```python
def acc_cont(df, y_):
    ct = pd.crosstab(df.shot_made_flag, df[y_]).apply(lambda x:x/x.sum(), axis=0)
    x, y = ct.columns, ct.values[1, :]
    plt.figure(figsize=(7, 5))
    plt.plot(x, y, 'ob')
    plt.xlabel(y_)
    plt.ylabel('% shots made')
    plt.savefig(y_ + '_vs_accuracy.png')
```

```python
acc_cont(data, 'shot_distance') #Accuracy with shot distance
```

```python
acc_cont(data, 'minutes_remaining') #Accuracy with minutes remaining
```

```python
def acc_cat(df, y_):
    ct = pd.crosstab(df.shot_made_flag, data[y_]).apply(lambda x:x/x.sum(), axis=0)
    ct.loc[1].plot(kind='bar')
```

```python
season_year = []
for i in range(len(data)):
    season_year.append(data.season[i][0:4])
season = np.array(season_year)
text_cols = ['action_type', 'combined_shot_type','shot_type', 'shot_zone_area', 'shot_zone_basic', 'shot_zone_range','opponent']
numeric_cols = ['lat','loc_x', 'loc_y', 'lon', 'minutes_remaining', 'period', 'seconds_remaining', 'shot_distance', 'shot_made_flag']
vectorized_data = pd.DataFrame()
vectorized_data['shot_id'] = data.shot_id
vectorized_data['year'] = season.astype('int')
for i in numeric_cols:
    vectorized_data[i] = data[i]
for i in text_cols:
    x = pd.get_dummies(data[i])
    vectorized_data = vectorized_data.join(x, rsuffix = i)
```

```python
made_shots = vectorized_data[vectorized_data.shot_made_flag == 1]
a = made_shots.groupby('year')
summ = a.sum()
summ = summ.reset_index()
```

```python
missed_shots = vectorized_data[vectorized_data.shot_made_flag == 0]
b = missed_shots.groupby('year')
summ_missed = b.sum()
summ_missed = summ_missed.reset_index()
```

```python
#Made shots 2pts and 3pts per year and total
plt.plot(summ['year'].values, summ['2PT Field Goal'], 'or')
plt.plot(summ['year'].values, summ['3PT Field Goal'], 'ob')
plt.plot(summ['year'].values, summ['shot_made_flag'], 'og')
```

```python
#Missed shots 2pts and 3pts per year and total
plt.plot(summ_missed['year'].values, summ_missed['2PT Field Goal'], 'or')
plt.plot(summ_missed['year'].values, summ_missed['3PT Field Goal'], 'ob')
plt.plot(summ_missed['year'].values, summ_missed['2PT Field Goal']+summ_missed['3PT Field Goal'], 'og')
```

```python
#Made shots 2pts and 3pts per year and total
plt.plot(summ['year'].values, summ['2PT Field Goal']/len(vectorized_data[vectorized_data['2PT Field Goal'] == 1]), 'or')
plt.plot(summ['year'].values, summ['3PT Field Goal']/len(vectorized_data[vectorized_data['3PT Field Goal'] == 1]), 'ob')
```

```python
data2 = vectorized_data[vectorized_data['shot_made_flag'].isnull() == False]
validation = vectorized_data[vectorized_data['shot_made_flag'].isnull() == True]
X_valid = validation.drop(['shot_made_flag'], axis=1)
X = data2.drop(['shot_made_flag'], axis=1)
y = data2['shot_made_flag']
X_train, X_test, y_train, y_test = cv.train_test_split(X, y, test_size=0.30)
```

```python
LogReg = LM.LogisticRegression()
LRfit = LogReg.fit(X_train, y_train)
LRpred = LRfit.predict(X_test)
met.accuracy_score(LRpred, y_test)
LRpredprob = LRfit.predict_proba(X_valid)
```

```python
RF = ensemble.RandomForestClassifier()
RFfit = RF.fit(X_train, y_train)
RFpred = RFfit.predict(X_test)
met.accuracy_score(RFpred, y_test)
RFpredprob = RFfit.predict_proba(X_valid)
```
